## Basic queries in problem statement
##### * How many machines have more than 100 seats?

##### * How the number of tickets booked and total amount earned changed with time?

```python
tickets=pd.read_sql_query("""select * from tickets t
                     join bookings b on t.book_ref = b.book_ref """, connection)

tickets['book_date']=pd.to_datetime(tickets['book_date'])
tickets['date']=tickets['book_date'].dt.date
x=tickets.groupby('date')[['date']].count()
plt.figure(figsize=(18,6))
plt.plot(x.index, x['date'], color = "red", marker='*')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Number of Tickets', fontsize = 20)
plt.grid('b')        
plt.show()
```

```python
bookings=pd.read_sql_query("""select * from bookings""", connection)

bookings['book_date']=pd.to_datetime(bookings['book_date'])
bookings['date']=bookings['book_date'].dt.date
x=bookings.groupby('date')[['total_amount']].count()
plt.figure(figsize=(18,6))
plt.plot(x.index, x['total_amount'], color = "red", marker='*')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Total amount earned', fontsize = 20)
plt.grid('b')        
plt.show()
```

### Occupancy rate analysis 
* Calculating  the total revenue per year and average revenue per ticket for each aircraft.

```python
pd.read_sql_query("""select aircraft_code, ticket_count, total_revenue, total_revenue/ticket_count as avg_revenue_per_ticket
                    from(select aircraft_code, count(*) as ticket_count, sum(amount) as total_revenue
                    from ticket_flights tf
                    join flights f on tf.flight_id = f.flight_id
                    group by aircraft_code)""", connection)
```

#### Calculating average occupancy per aircraft.

```python
'''occupancy_rate = pd.read_sql_query("""select x.aircraft_code, avg(x.seats_count) as booked_seats, y.num_of_seats, avg(x.seats_count)/y.num_of_seats as occupancy_rate
                    from (select aircraft_code, f.flight_id, count(*) as seats_count
                    from boarding_passes bp
                    inner join flights f on bp.flight_id=bp.flight_id
                    group by aircraft_code, f.flight_id) as x 
                    inner join (select aircraft_code, count(*) as num_of_seats
                    from seats
                    group by aircraft_code) as y
                    on x.aircraft_code = y.aircraft_code 
                    group by x.aircraft_code""", connection)'''


occupancy_rate =pd.read_sql_query ("""SELECT x.aircraft_code, AVG(x.seats_count) AS booked_seats,
                                        y.num_of_seats, AVG(x.seats_count) / y.num_of_seats AS occupancy_rate
                                    FROM
                                        (SELECT
                                        aircraft_code,
                                        f.flight_id,
                                        COUNT(*) AS seats_count
                                    FROM
                                        boarding_passes bp
                                    INNER JOIN flights f ON bp.flight_id = f.flight_id
                                    GROUP BY
                                        aircraft_code, f.flight_id) AS x
                                    INNER JOIN
                                    (SELECT
                                        aircraft_code,
                                        COUNT(*) AS num_of_seats
                                    FROM
                                        seats
                                    GROUP BY
                                    aircraft_code) AS y
                                    ON x.aircraft_code = y.aircraft_code
                                    GROUP BY
                                    x.aircraft_code""", connection)
occupancy_rate
```

#### How much total revenue would increase, if all aircrafts' occupancy increased by 10%.

```python
occupancy_rate['Increased_occ_rate'] = occupancy_rate['occupancy_rate']+occupancy_rate['occupancy_rate']*0.1
occupancy_rate
```

## Conclusion :
In summary, analysing revenue data such as total revenue per year, average revenue per ticket and average load factor per aircraft is crucial for airlines looking to maximise their profitability. By analysing these indicators, airlines can identify areas for improvement and adjust their pricing and routing plans. A higher occupancy rate is an important feature that can increase profitability as it allows airlines to maximise revenue while minimising the costs associated with unoccupied seats. Airlines should revise the prices for each aircraft as the low and high prices are also the reason why people do not buy tickets for these aircraft. They should set a reasonable price based on the condition and features of the aircraft, which should be neither too low nor too high.

Furthermore, increasing occupancy rates should not come at the price of consumer satisfaction or safety. Airlines must strike a balance between the need to make a profit and the importance of providing quality service and complying with safety regulations. Airlines can succeed in the long term in a highly competitive business if they adopt a data-driven strategy to analyse and optimise yield.
